```python
def amount_iter(compl, max_steps, threshold_inf: int = 10):
    c = compl
    z = 0
    i = 0
    while (z*z.conjugate()).real < threshold_inf and i < max_steps:
        z = z**2 + c
        i += 1
    return i
```
